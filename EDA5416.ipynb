IMPORTING LIBRARIES


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

READ THE DATASET('TITANIC')

```python
data = pd.read_csv("tested.csv")
```

LINE PLOT

```python
sns.lineplot(data=data,x="Age",y="Embarked")
plt.title("LINE PLOT")
plt.show()
```

HEAT-MAP-(NULL VALUES)

```python
sns.heatmap(data.isnull())
plt.title("HEAT MAP")
plt.show()
```

COUNT PLOT

```python
sns.set_style("darkgrid")
sns.countplot(x="Survived",hue="Sex",data=data,palette="BuGn")
```

```python
sns.set_style("darkgrid")
sns.countplot(x="Survived",hue="Sex",data=data,palette="cubehelix")
```

```python
sns.countplot(x="SibSp",data=data,palette="Blues_r")
```

DISTPLOT

```python
sns.distplot(data['Age'],color="darkviolet",bins=10)
```

BAR PLOT

```python
sns.barplot(data=data,x="Age",y="Embarked")
plt.title("BAR PLOT")
plt.show()
```

BOX PLOT

```python
plt.figure(figsize=(10,8))
sns.boxplot(x="Pclass",y="Age",data=data,palette="Dark2_r")
```

SCATTER PLOT

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='Pclass',y='Age',data=data)
plt.title("SCATTER PLOT")
plt.show()
```

CAT PLOT

```python
sns.catplot(data=data,x='Age',y='Pclass')
plt.title("CAT PLOT")
plt.show()
```
